#### Question 1.1 (15 marks)


In this Question, we want to write a function for string compression. Given a string S, for any substring s of length d composed of the same character c, we want to compress s into a substring of length 2 in the form of ```cd```. For instance, given a string ```'aaabb!%%%**'```, it will be compressed into ```'a3b2!1%3*2'```.  In particular, if the character c is a digit, i.e., $c\in \{0,1,2,3,\ldots,9\}$, it will be compressed into ```**cd```. For instance, if the string is ```aaabb3333333333****555```, it will be compressed into ```a3b2**310*4**53```. 





- Your Pseudo-code here





Declare a dictionary called 'dict'
Declare a string called 'result'

For loop through dic
 Add '+1' if dic[x] is the same as the value of 'x' in S
 If not, then just put 1
 
for loop key, value in dic.items():
    concatenate str(key) and str(value)
 

````python
# Your code here

def Compression(S):
    dic = {}
    result = ""
    
    for x in S:
#         print(x)
        if x in dic:
#             print(x)
            dic[x] = dic[x] + 1
        else:
            dic[x] = 1
    
    for key, value in dic.items():
        result += str(key) + str(value)

#     print(result)
````

````python
# Test code

Compression('aaabb3333333333****555')
````
